#### a. non-normalized tfxidf weights

```python
def tfidf(data, logbase):
    #the IDF values for each term as a single list
    print("the IDF values for each term as a single list")
    N = len(data.columns)
    idf = []
    for i in range(len(data)):
        nk, result = 0,0
        for j in range(len(data.columns)):
            if data.iloc[i,j] != 0:
                nk += 1
            else:
                continue
        result = math.log2(N/nk)
        idf.append(result)

    idfTable = pd.DataFrame(idf, columns = ['IDF'], index = ['T1', 'T2', 'T3', 'T4', 'T5', 'T6', 'T7', 'T8', 'T9', 'T10', 'T11','T12','T13','T14','T15'])
    display(round(idfTable,2))
    
    #the resulting document-term matrix with the new weights as final output
    print("the resulting document-term matrix with the new weights as final output")
    weightedmatrix = data.apply(lambda x: np.asarray(x) * np.asarray(idf))
    weightedmatrix = weightedmatrix.iloc[:,0:10]
    display(round(weightedmatrix,2))
            
```

#### b. signal-to-noise ratio

```python
def signaltonoise(data, base):
    #get tfk for each row
    data['tfk'] = data.iloc[:,0:10].sum(axis=1)
    
    #get probk
    probk = data.iloc[:,0:10].apply(lambda x: np.asarray(x) / np.asarray(data['tfk']))
    
    #get infoT
    infoT = probk.apply(getInfoT).replace(np.nan,0)
    
    #get avginfo
    infoT['avg-info'] = infoT.iloc[:,0:10].sum(axis=1)
    avgTable = pd.DataFrame(round(infoT['avg-info'],2), columns = ['avg-info'])
    print("entropy (Avg-Info) values for each term")
    display(avgTable)
    
    #get Noise
    infoT['noise'] = infoT['avg-info'].apply(lambda x: -x)
    
    #get Signal
    infoT['tfk'] = data['tfk']
    for i in range(len(data)):
        infoT['signal'] = np.log2(infoT['tfk']) + infoT['noise']
        
    signal = infoT['signal']
    
    #get final signal table
    print("new weights using the signal-to-noise ratio approach")
    signalTable = data.iloc[:,0:10].apply(lambda x: np.asarray(x) * np.asarray(signal))
    display(round(signalTable,2))
    
def getInfoT(x):
    np.seterr(divide = 'ignore')
    result = -x * np.log2(x)
    return result
```

#### c. Term Discrimination

```python
def TermDiscrimination(data):
    #get D*
    data['D*'] = data.mean(axis=1)
    
    dataT = data.T

    #GET Centroid, sim(Di, D*)
    count, centroid = 0, 0
    for i in range(len(dataT)-1):
        XY, X, Y = 0,0,0
        temp = 0
        for j in range(len(dataT.columns)):
            yvalue = dataT.iloc[-1, j]
            Y += yvalue ** 2
            if dataT.iloc[i, j] != 0:
                xvalue = dataT.iloc[i, j]
                XY += xvalue * yvalue
                X += xvalue ** 2
        temp = XY/(math.sqrt(X) * math.sqrt(Y))
        centroid+=temp
        count += 1
    
    avgSim = centroid/count
    print("AVG-SIM:", round(avgSim,3))
    
    #get simk when termk removed
    simlist = []
    for i in range(0, len(dataT.columns)):
        count, centroid = 0, 0
        for m in range(len(dataT)-1):
            XY, X, Y = 0,0,0
            temp = 0
            for n in range(len(dataT.columns)):
                if n != i:
                    yvalue = dataT.iloc[-1, n]
                    Y += yvalue ** 2
                    if dataT.iloc[m, n] != 0:
                        xvalue = dataT.iloc[m,n]
                        XY += xvalue * yvalue
                        X += xvalue ** 2
            temp = XY/(math.sqrt(X) * math.sqrt(Y))
            centroid+=temp
            count += 1
            
        avgSimk = centroid/count    
        simlist.append(avgSimk)
    
    #get disc table
    count = 0
    disctable = []
    for i in simlist:
        disctable.append(i-avgSim)
    
    finaldisctable = pd.DataFrame(disctable, columns = ['disck'], index = df.index)
    print("the average similarity across documents with that term omitted:")
    display(finaldisctable)
    
    print("identify the most and the list discriminating terms in the vocabulary:")
    finaldisctable = finaldisctable.sort_values(by=['disck'], ascending=False)
    display(finaldisctable)
    
    #term Discrimination table
    termDisc = data.apply(lambda x: np.asarray(x) * np.asarray(disctable))
    print("Term Weights Matrix:")
    display(termDisc)
```

```python
def DotProduct(data):
    Dot = []
    for i in range(len(data)-1):
        count = 0
        for j in range(len(data.columns)):
            temp = data.iloc[i,j] * data.iloc[-1,j]
            count += temp
        Dot.append(count)
        
    DotTable = pd.DataFrame(Dot, columns = ['DotProduct'], index = ['D1', 'D2', 'D3', 'D4', 'D5', 'D6', 'D7', 'D8', 'D9', 'D10'])
    display(DotTable)
```

```python
def CosineSimilarity(data):
    Cos = []
    for i in range(len(data)-1):
        ytemp = 0
        xytemp = 0
        xtemp = 0
        for j in range(len(data.columns)):
            Yvalue = data.iloc[-1,j]
            y = Yvalue**2
            ytemp += y
            Xvalue = data.iloc[i,j]
            if Xvalue != 0:
                xy = Yvalue*Xvalue
                x = Xvalue**2
                xytemp += xy
                xtemp += x
        temp = xytemp/math.sqrt(xtemp*ytemp)
        Cos.append(temp)
    
    CosTable = pd.DataFrame(Cos, columns = ['CosineSimilarity'], index = ['D1', 'D2', 'D3', 'D4', 'D5', 'D6', 'D7', 'D8', 'D9', 'D10'])
    display(CosTable)
```

```python
def DiceCoef(data):
    Dice = []
    for i in range(len(data)-1):
        ytemp = 0
        xytemp = 0
        xtemp = 0
        for j in range(len(data.columns)):
            Yvalue = data.iloc[-1,j]
            y = Yvalue**2
            ytemp += y
            Xvalue = data.iloc[i,j]
            if Xvalue != 0:
                xy = Yvalue*Xvalue
                x = Xvalue**2
                xytemp += xy
                xtemp += x
        temp = (2*xytemp)/(xtemp+ytemp)
        Dice.append(temp)
    
    DiceTable = pd.DataFrame(Dice, columns = ['DiceCoef'], index = ['D1', 'D2', 'D3', 'D4', 'D5', 'D6', 'D7', 'D8', 'D9', 'D10'])
    display(DiceTable)
```

```python
def JaccardCoef(data):
    Jaccard = []
    for i in range(len(data)-1):
        ytemp = 0
        xytemp = 0
        xtemp = 0
        for j in range(len(data.columns)):
            Yvalue = data.iloc[-1,j]
            y = Yvalue**2
            ytemp += y
            Xvalue = data.iloc[i,j]
            if Xvalue != 0:
                xy = Yvalue*Xvalue
                x = Xvalue**2
                xytemp += xy
                xtemp += x
        temp = xytemp/(xtemp+ytemp-xytemp)
        Jaccard.append(temp)
    
    JaccardTable = pd.DataFrame(Jaccard, columns = ['DiceCoef'], index = ['D1', 'D2', 'D3', 'D4', 'D5', 'D6', 'D7', 'D8', 'D9', 'D10'])
    display(JaccardTable)
```
